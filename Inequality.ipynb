```python
import pandas as pd    #pandas dataframes
import matplotlib.pyplot as plt

data = pd.read_csv('happyscores_income.csv')     #read file

data
```

```python
income = data['avg_income']  #pull out income data
inequality = data['income_inequality']
satisfaction = data['avg_satisfaction']

plt.scatter(income, inequality, s=satisfaction * 30, alpha = 0.25)

plt.xlabel('Income ($)')
plt.ylabel('Inequality')
```

```python
avg_inequality = inequality.mean()
max_inequality = inequality.max()

print(avg_inequality, max_inequality)
```

## Sort Data

```python
data.sort_values('income_inequality', inplace = True)
```

```python
highest_inequality = data[data['income_inequality'] > 39] 
highest_inequality
```

```python
highest_inequality.iloc[0]  #the country with the highest inequality
```

```python
plt.scatter(highest_inequality['avg_income'], highest_inequality['income_inequality'], alpha = 0.5)

plt.xlabel('Income ($)')
plt.ylabel('Inequality')

plt.text(highest_inequality.iloc[0]['avg_income'],
         highest_inequality.iloc[0]['income_inequality'],
         highest_inequality.iloc[0]['country'])

plt.text(highest_inequality.iloc[-1]['avg_income'],
         highest_inequality.iloc[-1]['income_inequality'],
         highest_inequality.iloc[-1]['country'])


```

```python
highest_income = highest_inequality[::]

highest_income.sort_values('avg_income', inplace = True)

print(highest_income)
```

```python
plt.scatter(highest_income['avg_income'], highest_income['income_inequality'], alpha = 0.5)

plt.xlabel('Income ($)')
plt.ylabel('Inequality')

plt.text(highest_income.iloc[0]['avg_income'],
         highest_income.iloc[0]['income_inequality'],
         highest_income.iloc[0]['country'])

plt.text(highest_income.iloc[-1]['avg_income'],
         highest_income.iloc[-1]['income_inequality'],
         highest_income.iloc[-1]['country'])
```
